```python
a=leocr.Minute
leocr.Minute.dtype
```

```python
plt.figure(figsize=(18,8))
plt.style.use('fivethirtyeight')
ax = sns.countplot('Minute',hue='Player',data=leocr,palette='dark')
ax.set_xlabel(xlabel='Playing_Position',fontsize=20)
ax.set_ylabel(ylabel='Result',fontsize=35)
ax.set_title(label='comparison of position and players',fontsize=50)
plt.show()
```

```python
plt.figure(figsize=(15,8))
ax=sns.countplot(x='Result',data=leocr,palette='bone')
ax.set_title(label='Count of players on basis of heights',fontsize=50)
ax.set_xlabel(xlabel='height is foot per inch ',fontsize=20)
ax.set_ylabel(ylabel='Player',fontsize=30)
plt.show()
```
